### 数据预处理

```python
from PIL import Image
import os

dateSet_Path=os.path.join(os.getcwd(),'dataSet','PetImages')
def clean_invalid_ImageFile(Parent_Path):
    son_paths=["Cat","Dog"]
    for son_path in son_paths:
        image_dir=os.path.join(Parent_Path,son_path)
        for image_file in os.listdir(image_dir):
            ImageFile_path=os.path.join(image_dir,image_file)
            # if not image_file.lower().endswith((".jpg",".png",".jpeg")):
            #     os.remove(ImageFile_path)#删除非图片文件
            try:
                with Image.open(ImageFile_path) as img:
                    img.verify()
            except Exception :
                print("路径为："+str(ImageFile_path)+"的图片错误，准备删除")
                os.remove(ImageFile_path)
                print("删除成功")

clean_invalid_ImageFile(dateSet_Path)#清理图片

All_samples_with_Index=[]#(图片路径，标签)
def imagePath_with_Label(parent_Path):
    label_dict={"Cat":0,"Dog":1}
    image_SonDirs=["Cat","Dog"]
    for image_SonDir in image_SonDirs:
        image_Dir=os.path.join(parent_Path,image_SonDir)
        for image_file in os.listdir(image_Dir):
           All_samples_with_Index.append((os.path.join(image_Dir,image_file),label_dict[image_SonDir]))
imagePath_with_Label(dateSet_Path)#图片路径有标签
```
